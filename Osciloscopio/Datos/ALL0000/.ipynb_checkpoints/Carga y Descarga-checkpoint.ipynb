```python
doc1 = np.genfromtxt("F0000CH1.CSV",delimiter=",")
x = doc1[:,3]
y = doc1[:,4]

plt.figure(figsize=(15,10))
plt.plot(x,y,"o",ms=2)
plt.xlim(-0.1,0.1)
plt.xlabel(u"Tiempo(s)",fontsize=25)
plt.ylabel(u"Voltaje(V)",fontsize=25)
plt.show()
```
