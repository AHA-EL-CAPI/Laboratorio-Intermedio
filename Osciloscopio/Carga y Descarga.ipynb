```python
%pylab inline
from scipy.optimize import curve_fit
```

```python
doc0 = np.genfromtxt("F0000CH1.CSV",delimiter=",")
x = doc0[:,3]
y = doc0[:,4]

y = y[x > -0.1]
x = x[x > -0.1]

y = y[x < 0.1]
x = x[x < 0.1]

plt.figure(figsize=(15,10))
plt.plot(x,y,"o",ms=2)
plt.xlabel(u"Tiempo(s)",fontsize=25)
plt.ylabel(u"Voltaje(V)",fontsize=25)
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
plt.savefig("carga1.jpg")
plt.show()
```

```python
y = y[x<0.]
x = x[x<0.]

a = min(y)*63./100.
print a

def func(t,c,b):
    return -c*np.exp(-t/b)

nonlfit=curve_fit(func, x, y,p0=(10.,1.))
th=np.linspace(-0.1,0.,100000)

print nonlfit[0][1]

plt.figure(figsize=(15,10))
plt.plot(th,func(th,nonlfit[0][0],nonlfit[0][1]))
plt.plot(x,y,"o",ms=2)
plt.hlines(a,min(x),max(x),"r",label=u'67% de la carga máxima')
plt.xlabel(u"Tiempo(s)",fontsize=25)
plt.ylabel(u"Voltaje(V)",fontsize=25)
plt.legend(loc='lower right',fontsize=25)
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
plt.savefig("carga111.jpg")
plt.show()
```

```python
doc3 = np.genfromtxt("F0003CH1.CSV",delimiter=",")
x = doc3[:,3]
y = doc3[:,4]

y = y[x > 0.]
x = x[x > 0.]

b = (diff(sign(diff(y[y>10.]))) > 0).nonzero()[0] + 1
c = (diff(sign(diff(y[y>10.]))) < 0).nonzero()[0] + 1

plt.figure(figsize=(15,10))
plt.plot(x,y,"o",ms=2.5,label=u'Datos')
plt.xlabel(u"Tiempo(s)",fontsize=25)
plt.ylabel(u"Voltaje(V)",fontsize=25)

from scipy.optimize import curve_fit

def func(t,a,b,c,d,e):
    return a*np.exp(-b*t)*np.sin(c*t+d)+e

nonlfit=curve_fit(func, x, y,p0=(0.,0,10000.,0.,0))
th=np.linspace(0.,0.0001,100000)
plt.plot(th,func(th,nonlfit[0][0],nonlfit[0][1],nonlfit[0][2],nonlfit[0][3],nonlfit[0][4]),"r",lw=1.5,label='Fit')
print nonlfit[0][0],nonlfit[0][1],nonlfit[0][2],nonlfit[0][3],nonlfit[0][4]
print np.sqrt(np.diag(nonlfit[1]))
plt.legend(loc='best',fontsize=25)
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
plt.savefig("amorti1.jpg")
plt.show()
```
