```python
%pylab inline
from scipy import stats
```

```python
x = np.array([8.4,9.6,13.0,13.2,14.5,16.0])
y = np.array([667.815,587.562,501.567,471.314,447.148,438.793])
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
y_reg = slope*x + intercept
plt.figure(figsize=(15,10))
plt.plot(x,y,'o',label = u'Experimento',ms=10)
plt.plot(x,y_reg,ls='dashed',label= u'Regresión')
plt.title(u'$Longitud\ de\ onda\ \lambda \ vs.\ Posición\ de\ la\ línea$',fontsize=30)
plt.ylabel(u'$\lambda \ (nm)$',fontsize=30)
plt.xlabel(u'$Posición\ de\ la\ línea\ espectral\ (adimensional)$',fontsize=30)
plt.legend(loc='best',fontsize=30)
plt.xlim(8,16.5)
plt.savefig('regresion.jpg')
plt.show()
```

```python
slope, intercept,std_err,r_value**2
```

```python
x = np.array([8.4,13.7,15.5])
y= 1./(slope*x + intercept)/10**(-9)
n_ini = np.array([3.,4.,5.])
x_new = 1./4. - 1./(n_ini**2)
slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(x_new,y)
y_new = slope1*x_new + intercept1
plt.figure(figsize=(15,10))
plt.plot(x_new,y,'o',label = u'Experimento',ms=10)
plt.plot(x_new,y_new,ls='dashed',label= u'Regresión')
#plt.savefig('regresion.jpg')
plt.show()
```
