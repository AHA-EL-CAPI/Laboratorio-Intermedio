```python
%pylab inline
from scipy import stats
```

```python
def func(x,m,b):
    return m*x + b

V = np.array([110.,120.,130.,140.,150.,160.,170.,180.,190.,200.,210.,220.,230.,240.,250.,260.,270.])
i2 = np.array([2.60,2.73,2.79,2.83,2.96,3.06,3.17,3.28,3.38,3.46,3.60,3.68,3.73,3.79,3.90,3.98,4.05])
i3 = np.array([1.44,1.73,1.17,1.86,1.87,1.99,2.03,2.09,2.15,2.21,2.28,2.33,2.40,2.46,2.50,2.56,2.62])
i4 = np.array([1.11,1.21,1.30,1.35,1.39,1.41,1.47,1.53,1.58,1.63,1.68,1.72,1.77,1.81,1.84,1.89,1.93])
i5 = np.array([0.85,0.92,0.96,1.02,1.08,1.13,1.17,1.22,1.25,1.30,1.33,1.36,1.40,1.43,1.46,1.49,1.53])

B2 = ((4./5.)**(3./2.))*(4.*np.pi*10**(-7))*154.*i2/0.2
B3 = ((4./5.)**(3./2.))*(4.*np.pi*10**(-7))*154.*i3/0.2
B4 = ((4./5.)**(3./2.))*(4.*np.pi*10**(-7))*154.*i4/0.2
B5 = ((4./5.)**(3./2.))*(4.*np.pi*10**(-7))*154.*i5/0.2

x2 = ((B2*0.02)**2.)/2.
x3 = ((B3*0.03)**2.)/2.
x4 = ((B4*0.04)**2.)/2.
x5 = ((B5*0.05)**2.)/2.

x = np.append(x2,x3)
x = np.append(x,x4)
x = np.append(x,x5)

V_new = np.append(V,V)
V_new = np.append(V_new,V)
V_new = np.append(V_new,V)
slope, intercept, r_value, p_value, std_err = stats.linregress(x,V_new)
```
